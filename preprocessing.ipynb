```python
import pandas as pd
```

```python
with open('/team/courses/MSiA400/Jobs/ex140531_users_MYJOBS2_20140601001500.log', 'r') as f:
    log_file = f.readlines()
```

```python
log_file[0:10]
```

The fields of the log are in the fourth row of file

```python
fields = log_file[3].split()
fields
```

The first entry just denotes it's a field row and is not actually a column

```python
fields = fields[1:]
fields
```

The rows of data begin in the fifth row of the file

```python
data = log_file[4:]
```

```python
for i, row in enumerate(data):
    l = row.split()
    if len(l) < 18:
        l.insert(10, '-')
    data[i] = l
```

```python
df = pd.DataFrame(data, columns=fields)
```

```python
df.head(30)
```

```python
# Null fields do to incorrect parsing
pd.isnull(df['time-taken']).sum()
```

```python
# Total number of rows
pd.isnull(df['time-taken']).count()
```

```python
# Splits out uri query into id fields
df[['jvguid', 'buid', 'vsid', 'aguid']] = df['cs-uri-query'].str.split('&', expand=True).iloc[:, :4]
```

```python
# Remove type of id from field so all that's remaining is the actual id
df['jvguid'] = df['jvguid'].str.replace(r'.*=', '')
df['buid'] = df['buid'].str.replace(r'.*=', '')
df['vsid'] = df['vsid'].str.replace(r'.*=', '')
df['aguid'] = df['aguid'].str.replace(r'.*=', '')
```
